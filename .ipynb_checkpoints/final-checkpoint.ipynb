This project will examine Indian crime data, which can be found here: https://www.kaggle.com/datasets/meruvulikith/crimes-in-india-dataset-2001-2013. In this notebook I will attempt to answer the following questions: 

* How have the occurences of certain crimes changed over time? Have they changed differently for different crimes and do these changes differ by region?
* Do different regions have a dominant type of crime being committed? If they do, what is it and how has it changed over time? Why might this be?
* How have the rates of violent crime changed over time and by district?
* Have the rates of violent crimes against women changed over time? 

I start by importing the necessary libraries, loading our data, and examining how the data is formatted.

**Note:** It would be beneficial to review crime rates rather than number of incidents, however data for population by region or by district over time is not available to me and I cannot find a source of data that is suitable for this situation. Hence, I will be looking at number of occurences rather than rates. This means that some increases in crimes being commited are likely due to increases in population that we are not accounting for.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('https://raw.githubusercontent.com/calebchou21/DSCI235-Final/main/india_crime.csv')

num_unique_districts = data['DISTRICT'].nunique()
print("Number of unique districts:", num_unique_districts)

num_unique_states = data['STATE/UT'].nunique()
print("Number of unique states:", num_unique_states)

data.head()
```

Next, we can visualize the occurences of different categories of crime in order to get a better sense of our data.

```python
temp_data = data.drop(columns=['STATE/UT', 'DISTRICT', 'YEAR', 'TOTAL IPC CRIMES', 'OTHER IPC CRIMES'])
temp_totals = temp_data.sum()

plt.figure(figsize=(12, 8))
sns.barplot(x=temp_totals.values, y=temp_totals.index, palette='viridis')
plt.title('Total Count of Crimes in India by Category')
plt.xlabel('Total Count')
plt.ylabel('Crime Categories')
plt.show()
```

```python
empty_labels = [""] * len(temp_totals.index)

# Hide very small percentages
def my_autopct(pct):
    return '{:.1f}%'.format(pct) if pct > 1 else ''

plt.figure(figsize=(10, 8))
plt.pie(temp_totals, labels=empty_labels, autopct=my_autopct, startangle=140, pctdistance=1.06)
plt.title('Total Count of Crimes in India by Category')
plt.axis('equal')

plt.legend(temp_totals.index, bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

Now, we have a better idea of the crimes which are being commited the most and which are being committed the least.

**How have the occurences of certain crimes changed over time?** In order to investigate this question I would first like to examine how the yearly incidents of several categories of crimes has changed over time. I quickly realized that there are too many categories to plot all at once, so I have decided to limit this investigation to the categories that I am most interested in. I begin by creating a line plot detailing the incidents of each of my selected categories of crime over time.

```python
# Group our data by year
grouped_data = data.groupby(['YEAR']).sum()

# Select crimes we would like to analyze + plot each crime
crime_categories = ['MURDER', 'RAPE', 'THEFT', 'RIOTS', 'CHEATING', 'HURT/GREVIOUS HURT']
plt.figure(figsize=(10, 6))
for crime_category in crime_categories:
    plt.plot(grouped_data.index, grouped_data[crime_category], label=crime_category)

plt.title('Trend of Crime Incidents Over Time')
plt.xlabel('Year')
plt.ylabel('Number of Incidents')
plt.legend()
plt.grid(True)
plt.show()
```

**Discussion of Plot:** Focussing on Theft and Hurt/Grevious Hurt, we can see that these categories of crimes occur at rates significantly higher than those of the other categories being investigated, and have also risen at an unprecedented rate. Unfortunately we do not have data regarding poverty levels in India, as it would have been interesting to investigate a possible correlation. Interestingly, it seems there is evidence that poverty in India has actually consistently fallen over time (Sandefur et al., 2022). Incidents of murder remain consistent over time, with the crime of cheating consistently rising. 

**Does the rate of crime differ by region?** To investigate this question we will create seperate visualizations in order to get a better sense of whether or not each of these categories of crime changes in different ways by region. As we cannot visualize all of the regions here, I select the regions we should investigate by finding the regions with the lowest and highest incidents of each crime category in a year. Additionally, I ensure that the selected regions have more than 5 recorded years of data.

```python
# Group the data by region and count years of data
grouped_data = data.groupby('STATE/UT')['YEAR'].nunique()

# Select regions with more than five years of data
regions_with_enough_data = grouped_data[grouped_data > 5].index

# Initialize sets to store the names of regions with highest and lowest crime occurrences
highest_crime_regions = set()
lowest_crime_regions = set()

# Identify regions with lowest and highest crime incidents in each category
for crime_category in crime_categories:
    crime_data_filtered = data[data['STATE/UT'].isin(regions_with_enough_data)]
    highest_region = crime_data_filtered.groupby('STATE/UT')[crime_category].sum().idxmax()
    lowest_region = crime_data_filtered.groupby('STATE/UT')[crime_category].sum().idxmin()
    highest_crime_regions.add(highest_region)
    lowest_crime_regions.add(lowest_region)

regions_to_investigate = highest_crime_regions | lowest_crime_regions

print("Regions to investigate:")
for region  in regions_to_investigate:
    print(f"- {region}")
```

Now, lets create some plots that can visualize trends in incidents of crimes over time seperated by region.

```python
from matplotlib.gridspec import GridSpec

# This configures our grid layout for our plots
num_plots = len(regions_to_investigate)
cols = 2
rows = (num_plots + 1) // cols
height_ratios = [.8] * rows
fig = plt.figure(figsize=(15, 5 * rows))
gs = GridSpec(rows, cols, height_ratios=height_ratios)

# Plot the trends for each crime category
for i, region in enumerate(regions_to_investigate):
    row = i // cols
    col = i % cols
    ax = fig.add_subplot(gs[row, col])
    
    region_data = data[data['STATE/UT'] == region].groupby('YEAR').sum()
    
    for crime_category in crime_categories:
        ax.plot(region_data.index, region_data[crime_category], label=crime_category)
        
    ax.set_title(f'{region}')
    ax.set_xlabel('Year')
    ax.set_ylabel('Number of Incidents')
    ax.legend()

plt.tight_layout()
plt.show()
```

**Discussion of Plots:** These plots paint a much clearer picture! We can see that crime trends differ by region quite significantly. Although the Lakshadweep region plot appears to be very turbulent in it's trends, the y-axis is scaled incredibly low compared to these other regions. Lakshadweep is a small touristy area of India, although the spikes in the occurences of rioting are potentially related to past protests against government policy in this area (Reuters, 2021). Additionally interesting is the rapid uptick in crimes of cheating in Rajasthan and relative unusualness of the trends in crime occurences for the different categories. The Panjab region also has unusually high cases of cheating and a potentially strange uptick in theft and hurt/grievious hurt.

**Do different regions have a dominant type of crime being committed?** We have now answered our next question! It seems that different regions do have different dominant types of crime being committed, however, it is almost always either theft or hurt/grievous hurt (I believe this is akin to assualt). This makes sense as the majority of the regions we are looking at are rather under-developed and struggle with poverty and these are relatively low stakes crimes. Next we will look at the more serious crimes, where I imagine we will see these differences are more predominant. 

**How have the rates of violent crime changed over time?** To answer this question lets first create a plot to visualise the trends in these violent crimes across all of India.

```python
grouped_data = data.groupby(['YEAR']).sum()
# List of violent crimes types we will investigate
violent_crime_categories = ['MURDER', 'ATTEMPT TO MURDER', 'CULPABLE HOMICIDE NOT AMOUNTING TO MURDER', 
                            'OTHER RAPE', 'KIDNAPPING & ABDUCTION']

plt.figure(figsize=(10, 6))
for crime_category in violent_crime_categories:
    plt.plot(grouped_data.index, grouped_data[crime_category], label=crime_category)

plt.title('Trend of Crime Incidents Over Time')
plt.xlabel('Year')
plt.ylabel('Number of Incidents')
plt.legend()
plt.grid(True)
plt.show()
```

**Discussion of Plot:** The most notable thing here is the strangely large increase in the occurences of kidnapping and abduction. It seems that the occurences of murder have reamined fairly steady, with other categories of crime slowly increasing, possibly on account of population growth.

Next lets examine how these trends in violent crimes look within districts. With over 800 unique districts in our dataset, it would be difficult to visually examine all of them, so lets find the district with the highest and lowest changes in occurences of crimes.

```python
grouped_data = data.groupby(['DISTRICT', 'YEAR'])[violent_crime_categories].sum()
# Filter these out
grouped_data = grouped_data[grouped_data.index.get_level_values('DISTRICT') != 'TOTAL']
grouped_data = grouped_data[grouped_data.index.get_level_values('DISTRICT') != 'DELHI UT TOTAL']

# Calculate the difference in total crime incidents between the first and last available years for each district
districts_rate_of_change = {}
for district, group in grouped_data.groupby(level=0):
    min_year = group.index.get_level_values('YEAR').min()
    max_year = group.index.get_level_values('YEAR').max()
    first_year_count = group.loc[(district, min_year), :].sum()
    last_year_count = group.loc[(district, max_year), :].sum()
    change = last_year_count - first_year_count
    districts_rate_of_change[district] = change
    
# Find districts with highest and lowest changes in violent crime occurences
district_with_highest_change = max(districts_rate_of_change, key=lambda x: districts_rate_of_change[x])
district_with_lowest_change = min(districts_rate_of_change, key=lambda x: districts_rate_of_change[x])

print("District with the Highest Rate of Change in Total Crime Incidents Over Time:")
print(district_with_highest_change)
print("\nDistrict with the Lowest Rate of Change in Total Crime Incidents Over Time:")
print(district_with_lowest_change)
```

Now that we have found our districts of interest, lets again plot the trends in occurences of violent crimes.

```python
chosen_districts_data = data[data['DISTRICT'].isin([district_with_highest_change, district_with_lowest_change])]

# Group the filtered data by district, year, and violent crime category
grouped_data = chosen_districts_data.groupby(['DISTRICT', 'YEAR'])[violent_crime_categories].sum()

plt.figure(figsize=(16, 8))
for i, district in enumerate([district_with_highest_change, district_with_lowest_change]):
    plt.subplot(1, 2, i+1)
    for crime_category in violent_crime_categories:
        plt.plot(grouped_data.loc[district].index, grouped_data.loc[district][crime_category], label=crime_category)

    plt.title(f'Trends in Violent Crime Incidents - {district}')
    plt.xlabel('Year')
    plt.ylabel('Number of Incidents')
    plt.legend()
    plt.grid(True)

plt.tight_layout()
plt.show()
```

**Discussion of Plots:** These paint an interesting picture. On the left we can see rates of kidnapping, attempted murder, and sexual assault have skyrocketed over time in the Murshidabad district. On the other hand, in the Etah district, we have a fairly rapid decrease in attempted murder, murder, and kidnapping, although this must be a significntly smaller region. It seems that Murshidabad is a district that has had problems with terrorism and crime associated with that, which may be why we are seeing such an increase in these crimes (Majumder, 2020). 

**Have the rates of violent crimes against women changed over time?** I would like to conclude this notebook by attempting to investigate how crime rates against women have changed over time. India has had noteable problems with cases of violence against women in recent years (Pandey, 2022), so I would like to investigate whether this has been a consistent problem or whether it is newer. Lets create some plots to answer this question.

```python
"""

Seperated into two plots for better visibility of crimes with lower
occurences.

"""

grouped_data = data.groupby(['YEAR']).sum()
violent_crime_categories = ['OTHER RAPE', 'ASSAULT ON WOMEN WITH INTENT TO OUTRAGE HER MODESTY',\
                            'CRUELTY BY HUSBAND OR HIS RELATIVES']

# Setup grid
fig, axs = plt.subplots(1, 2, figsize=(15, 6))

# Plot for violent crime categories
for crime_category in violent_crime_categories:
    axs[0].plot(grouped_data.index, grouped_data[crime_category], label=crime_category)

axs[0].set_title('Trend of Violent Crime Incidents Over Time')
axs[0].set_xlabel('Year')
axs[0].set_ylabel('Number of Incidents')
axs[0].legend()
axs[0].grid(True)

# Plot for 'RAPE' and 'DOWRY DEATHS'
rape_data = grouped_data['RAPE']
dowry_deaths_data = grouped_data['DOWRY DEATHS']
axs[1].plot(rape_data.index, rape_data, label='RAPE', color='red')
axs[1].plot(dowry_deaths_data.index, dowry_deaths_data, label='DOWRY DEATHS', color='blue')
axs[1].set_title('Trend of Rape and Dowry Deaths Incidents Over Time')
axs[1].set_xlabel('Year')
axs[1].set_ylabel('Number of Incidents')
axs[1].legend()
axs[1].grid(True)
plt.tight_layout()
plt.show()
```

Here, we can see that these crimes have increased at an alarming rate between 2002 and 2012. Especially alarming are the rapidly increasing rates of 'cruelty by husband or his relatives' and rape.
